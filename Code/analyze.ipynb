```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv('datasets/train.csv')

# Count the instances of each class in the 'extent' column
extent_counts = data['extent'].value_counts()

# Calculate the percentage of each class
total_instances = data['extent'].count()
percentages = (extent_counts / total_instances) * 100

print("Class Counts:")
print(extent_counts)
print("\nPercentage of Dataset:")
print(percentages)
```

```python
plt.figure(figsize=(8, 6))
plt.pie(percentages, labels=extent_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Extent Class Distribution')
plt.show()

# Create a histogram for the frequencies
plt.figure(figsize=(8, 6))
plt.bar(extent_counts.index, extent_counts)
plt.xlabel('Extent Classes')
plt.ylabel('Frequency')
plt.title('Extent Class Frequencies')
plt.show()
```
